```python
from pathlib import Path
import pandas as pd
import gzip, json
# Locate the file even if it was moved: search cwd and parents
candidates = list(Path.cwd().rglob('ice+term-uitkeyword.jsonl.gz'))
if not candidates:
    candidates = []
    for base in [Path.cwd()] + list(Path.cwd().parents)[:5]:
        p = base / 'ice+term-uitkeyword.jsonl.gz'
        if p.exists():
            candidates.append(p)
if not candidates:
    raise FileNotFoundError('ice+term-uitkeyword.jsonl.gz not found; update path or move file back to repo')
p = candidates[0]
with gzip.open(p, 'rt', encoding='utf-8') as f:
    objs = [json.loads(l) for l in f]
df = pd.json_normalize(objs)

```
