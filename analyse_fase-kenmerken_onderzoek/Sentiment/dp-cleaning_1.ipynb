```python
from pathlib import Path
import pandas as pd
import gzip, json

# The notebook's directory as the base
notebook_dir = Path('../../aqcuisition_fase-scrapers/Debatepolitics/ice+term-uitkeyword.jsonl.gz')


with gzip.open(notebook_dir, 'rt', encoding='utf-8') as f:
    objs = [json.loads(l) for l in f]
df = pd.json_normalize(objs)
# print panda's version
print(f"Pandas version: {pd.__version__}")
```
