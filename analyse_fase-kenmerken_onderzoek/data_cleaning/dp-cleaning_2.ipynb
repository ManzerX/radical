```python
from pathlib import Path
import pandas as pd
import gzip, json
data_dir = Path("aqcuisition_fase-scrapers/Debatepolitics/data/debatepolitics/matches.jsonl.gz")

with gzip.open(data_dir, 'rt', encoding='utf-8') as f:
    objs = [json.loads(l) for l in f]
df = pd.json_normalize(objs)
df.info()
```

```python
df.isna().sum()
```

```python
df1 = df.drop(columns=['thread.id', 'thread.forum_url'], inplace=False)
df1.info()
```

```python
df1.to_json('analyse_fase-kenmerken_onderzoek/Sentiment/dp-set2-cleaned.jsonl.gz', orient='records', lines=True, compression='gzip')
df2 = pd.read_json('analyse_fase-kenmerken_onderzoek/Sentiment/dp-set2-cleaned.jsonl.gz', orient='records', lines=True, compression='gzip')
df2.info()
```
